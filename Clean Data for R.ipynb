```python
numerical_columns = df.dtypes[(df.dtypes == 'float64')].index.values
```

```python
numerical_columns
```

```python
#To remove ['SpecificGravityUrine', 'Albumin', 'Sugar'])
numerical_columns =[x for x in  list(['Age', 'BloodPressure', 'SpecificGravityUrine', 'Albumin', 'Sugar',
       'BloodGlucose', 'BloodUrea', 'SerumCreatine', 'Sodium',
       'Potassium', 'Hemoglobin', 'PackedCellVolume',
       'WhiteBloodCellCount', 'RedBloodCellCount']) if x not in list(['SpecificGravityUrine', 'Albumin', 'Sugar'])]
numerical_columns
```

```python
len(numerical_columns)
```

```python
from sklearn.preprocessing import MinMaxScaler

def scale_column(column):
    scaler = MinMaxScaler()
    result = scaler.fit_transform(pd.DataFrame(column))[:,0]
    print(result.shape)
    return result

df_scaled = df.copy()

df_scaled[numerical_columns] = df[numerical_columns].apply(scale_column, axis=0)

df_scaled.head()
```
